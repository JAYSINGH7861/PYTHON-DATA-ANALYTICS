```python
squares={1:1,2:4,3:9,4:16,5:25} # create a dict 
print(squares.pop(3))    # remove a particular items,return its value #output : 9
print(squares.popitem())  # remove an arbitrary items ,return (key,value)
squares.clear()  # remove all items 
print(squares)
```

```python
contacts={
    'police':112,
    'emergency':101,
}

while True:
    name=input('Search any contact:')
    if len(name)==0:
        print("bye")

    if name in contacts:
        print(f"{name}:{contacts[name]}")

```
